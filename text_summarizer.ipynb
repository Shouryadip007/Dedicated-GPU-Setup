```python
import torch
from transformers import pipeline

# 1. Check for GPU (Your 3050)
device = 0 if torch.cuda.is_available() else -1
print(f"Running on: {'GPU' if device == 0 else 'CPU'}")

# 2. Load the summarization model
# 'facebook/bart-large-cnn' is great for English articles
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

# 3. Your text
text = input("Give a Paragraph or note for summarize\n")

# 4. Generate Summary
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print("\n--- Summary ---")
print(summary[0]['summary_text'])
```
